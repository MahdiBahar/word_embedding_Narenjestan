```python
from hazm import sent_tokenize, word_tokenize, POSTagger
import stanza

# orginal_text = "تراکنش رمز دوم پویا بر بستر تراکنش نیازمند رمز دوم با آبونمان و بدون آبونمان انجام می‌گردد"

# 1) Hazm sentence + word tokenize
hazm_sents = [word_tokenize(s) for s in sent_tokenize(orginal_text)]
# Sanity: list of lists
# e.g., [['تراکنش','رمز',...,'می‌گردد']]

# 2) Hazm POS on the same tokens (flatten -> tag -> reshape)
tagger = POSTagger(model=model_hazm_path)
flat_tokens = [tok for sent in hazm_sents for tok in sent]
flat_tags   = tagger.tag(flat_tokens)  # list of (token, POS)
# reshape to per-sentence
idx = 0
hazm_tags_sents = []
for sent in hazm_sents:
    hazm_tags_sents.append(flat_tags[idx: idx+len(sent)])
    idx += len(sent)

# 3) Stanza with pretokenized input (NO mwt)
nlp = stanza.Pipeline(
    lang='fa',
    processors='tokenize,pos,lemma',
    tokenize_pretokenized=True,
    tokenize_no_ssplit=True,   # keep your sentence boundaries
    pos_batch_size=32,         # safer on RAM
    download_method=None,
    use_gpu=False
)

# 4) Run Stanza on your tokens
#    Pass list-of-sentences (list of list of strings)
doc = nlp(hazm_sents)

# 5) Collect lemmas in the same structure
stanza_words_sents = [sent.words for sent in doc.sentences]
stanza_lemmas_sents = [[w.lemma for w in words] for words in stanza_words_sents]

# 6) Replace only verbs per Hazm with Stanza lemma (keep alignment)
out_sents = []
for sent_tags, sent_lemmas in zip(hazm_tags_sents, stanza_lemmas_sents):
    out = []
    for (tok, pos), lemma in zip(sent_tags, sent_lemmas):
        out.append(lemma if "VERB" in pos and lemma else tok)
    out_sents.append(out)

# 7) Rebuild text
processed_text = " ".join(" ".join(s) for s in out_sents)
# Flatten both
out_tokens = [tok for sent in out_sents for tok in sent]

print("Original tokens (Hazm):", len(flat_tokens))
print("Out tokens (verbs lemmatized):", len(out_tokens))
print("Same length? ", len(flat_tokens) == len(out_tokens))
# print(processed_text)
```
