## using Stanza for tokenizing

```python
from hazm import POSTagger
import stanza

# ---- Input ----
# orginal_text = "تراکنش رمز دوم پویا بر بستر تراکنش نیازمند رمز دوم انجام می‌گردد"

# ---- Load models ----
tagger = POSTagger(model="resources/pos_tagger.model")
nlp = stanza.Pipeline(
    lang='fa',
    processors='tokenize,mwt,pos,lemma',
    download_method=None,
    use_gpu=False
)

# ---- Stanza pass: tokenize + lemma ----
doc = nlp(orginal_text)

# Flattened stanza words (keep order)
stanza_words = []          # list of word objects (to keep misc/space info)
stanza_tokens_text = []    # parallel list of word.text (for Hazm)
stanza_lemmas = []         # parallel list of word.lemma

for sent in doc.sentences:
    for w in sent.words:   # 'words' (not 'tokens') to align with lemmas
        stanza_words.append(w)
        stanza_tokens_text.append(w.text)
        stanza_lemmas.append(w.lemma)

# ---- Hazm POS on *Stanza tokens* ----
hazm_tags = tagger.tag(stanza_tokens_text)  # same length/order as stanza_words

# ---- Build output tokens: lemma only if Hazm says VERB ----
out_tokens = []
for (w, (tok_text, hazm_pos), lemma) in zip(stanza_words, hazm_tags, stanza_lemmas):
    if "VERB" in hazm_pos:
        out_tokens.append(lemma if lemma else tok_text)
    else:
        out_tokens.append(tok_text)

# ---- Reconstruct sentence respecting Stanza spacing ----
# Stanza usually marks 'SpaceAfter=No' in w.misc when no space should follow.
rebuilt = []
for i, (w, tok) in enumerate(zip(stanza_words, out_tokens)):
    rebuilt.append(tok)
    space_after_no = w.misc and "SpaceAfter=No" in w.misc
    if not space_after_no and i < len(stanza_words) - 1:
        rebuilt.append(" ")
lemmatized_sentence_stanza = "".join(rebuilt)
lemmatized_sentence_stanza1 = " ".join(out_tokens)

# ---- Sanity checks ----
print("len(stanza_tokens_text) =", len(stanza_tokens_text))
print("len(out_tokens)         =", len(out_tokens))
print("chars in output         =", len(lemmatized_sentence_stanza))

# print("Original:", orginal_text)
# print("Processed:", lemmatized_sentence_stanza)
# print("Processed1:", lemmatized_sentence_stanza1)
```

```python
from hazm import sent_tokenize, word_tokenize, POSTagger
import stanza

# orginal_text = "تراکنش رمز دوم پویا بر بستر تراکنش نیازمند رمز دوم با آبونمان و بدون آبونمان انجام می‌گردد"

# 1) Hazm sentence + word tokenize
hazm_sents = [word_tokenize(s) for s in sent_tokenize(orginal_text)]
# Sanity: list of lists
# e.g., [['تراکنش','رمز',...,'می‌گردد']]

# 2) Hazm POS on the same tokens (flatten -> tag -> reshape)
tagger = POSTagger(model="resources/pos_tagger.model")
flat_tokens = [tok for sent in hazm_sents for tok in sent]
flat_tags   = tagger.tag(flat_tokens)  # list of (token, POS)
# reshape to per-sentence
idx = 0
hazm_tags_sents = []
for sent in hazm_sents:
    hazm_tags_sents.append(flat_tags[idx: idx+len(sent)])
    idx += len(sent)

# 3) Stanza with pretokenized input (NO mwt)
nlp = stanza.Pipeline(
    lang='fa',
    processors='tokenize,pos,lemma',
    tokenize_pretokenized=True,
    tokenize_no_ssplit=True,   # keep your sentence boundaries
    pos_batch_size=32,         # safer on RAM
    download_method=None,
    use_gpu=False
)

# 4) Run Stanza on your tokens
#    Pass list-of-sentences (list of list of strings)
doc = nlp(hazm_sents)

# 5) Collect lemmas in the same structure
stanza_words_sents = [sent.words for sent in doc.sentences]
stanza_lemmas_sents = [[w.lemma for w in words] for words in stanza_words_sents]

# 6) Replace only verbs per Hazm with Stanza lemma (keep alignment)
out_sents = []
for sent_tags, sent_lemmas in zip(hazm_tags_sents, stanza_lemmas_sents):
    out = []
    for (tok, pos), lemma in zip(sent_tags, sent_lemmas):
        out.append(lemma if "VERB" in pos and lemma else tok)
    out_sents.append(out)

# 7) Rebuild text
processed_text = " ".join(" ".join(s) for s in out_sents)
# Flatten both
out_tokens = [tok for sent in out_sents for tok in sent]

print("Original tokens (Hazm):", len(flat_tokens))
print("Out tokens (verbs lemmatized):", len(out_tokens))
print("Same length? ", len(flat_tokens) == len(out_tokens))
# print(processed_text)
```
