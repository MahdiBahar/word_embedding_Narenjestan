## Insert Data from txt file

```python
tokens = []

# file_name = "tagged_lemmatized_tokens_before_compondverb_example"
# file_name = "tagged_pure_tokens_before_compondverb_example"
file_name = "tagged_tokens_example"
with open(file_name+".txt", encoding="utf-8") as file: #reading the text file
    
    for line in file: #reads every line of the text file
        if not line:
            continue #if we had an empty line...
        line = line.strip() #removes white spaces in our lines
        word, tag = line.split("\t") #split each line into words and tags
        tokens.append({'word': word, 'pos': tag})  #put this tupples into our list
```
